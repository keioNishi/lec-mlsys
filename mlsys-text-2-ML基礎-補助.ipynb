## 線形代数
線形代数は、複数の変数間の関係をシンプルに記述できるため、機械学習において頻出する。

### スカラー，ベクトル，行列，テンソル

#### スカラー

スカラーは，1つの値もしくは変数のことで、例えば以下のように表す。
$$
x, y, M, N
$$

#### ベクトル

ベクトルは、複数のスカラーを縦もしくは横方向に集めて並べたもので、以下のように表す。

$$
\boldsymbol{x}=\begin{bmatrix}
x_{1} \\
x_{2} \\
x_{3}
\end{bmatrix}, \
\boldsymbol{y}=\begin{bmatrix}
y_{1} \\
y_{2} \\
\vdots \\
y_{N}\end{bmatrix}
$$

ベクトルの表記は太文字とすることでスカラーかベクトルかを区別し、縦方向に並べたものを列ベクトル、横方向に並べたものを行ベクトルと呼ぶ。

一般に、単にベクトルと表現した場合には列ベクトルを指すことが多い。

#### 行列

行列は複数の同じサイズのベクトルを並べたもので、以下のように表す。

$$
\boldsymbol{X}=\begin{bmatrix}
x_{11} & x_{12} \\
x_{21} & x_{22} \\
x_{31} & x_{32}
\end{bmatrix}
$$

多くの場合、行列は大文字または大文字の太文字で表記する。



#### テンソル

テンソルはベクトルや行列を一般化した概念で、スカラーは0階、ベクトルは1階、行列は2階のテンソルと表現できる。

行列を奥行き方向にさらに並べた多次元配列を3階のテンソルと呼び、それ以上の多次元配列も同様に高階のテンソルと呼ぶ。

- 例えば、カラー画像をデジタル表現する場合、画像を構成する各ピクセルはRGBの色空間を用いるのが一般的で、「行・列・色（チャンネル）」の3つの軸（次元）を持つ3階のテンソルとして表現される。

機械学習（特にディープラーニング）のライブラリ（PyTorch, TensorFlowなど）において単に「テンソル」と表現されている場合は、スカラーから高次元配列までの総称を指す。

#### 転置

列ベクトルを行ベクトルに、行ベクトルを列ベクトルに入れ替える（あるいは行列の行と列を入れ替える）演算を**転置 (Transpose）** と呼び，$T$または$	op$で表記する。

$$
\begin{aligned}\boldsymbol{x}&=\begin{bmatrix}
1 \\
2 \\
3
\end{bmatrix}, \
\boldsymbol{x}^{T}=\begin{bmatrix} 1 & 2 & 3 \end{bmatrix} \\
\boldsymbol{X}&=\begin{bmatrix}
1 & 4 \\
2 & 5 \\
3 & 6
\end{bmatrix}, \
\boldsymbol{X}^{T}=\begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6
\end{bmatrix}\end{aligned}
$$

のようになる

行列に対する転置では、サイズが$(N, M)$から$(M, N)$となり、$i$行$j$列目の値が$j$行$i$列目の値になる。

転置に関して、次の重要な性質がある。

$$
\begin{aligned}
&(1) \ \left( \boldsymbol{A}^{T}\right)^{T}=\boldsymbol{A}\\
&(2) \ \left( \boldsymbol{A}\boldsymbol{B}\right) ^{T}=\boldsymbol{B}^{T}\boldsymbol{A}^{T}\\
&(3) \ \left( \boldsymbol{A}\boldsymbol{B}\boldsymbol{C}\right) ^{T}=\boldsymbol{C}^{T}\boldsymbol{B}^{T}\boldsymbol{A}^{T}
\end{aligned}
$$


### 線形結合と二次形式

単位行列と逆行列については、既存の知識を前提としてここでは詳細を省略する。
なお、機械学習では巨大な行列を扱うが、逆行列を直接計算することは計算コストや数値的安定性の面から避けられることが多く、連立一次方程式の解法アルゴリズムや反復法が用いられる。実装にはライブラリ（NumPy, PyTorchなど）の利用が不可欠である。

機械学習では、$\boldsymbol{b}^{T}\boldsymbol{x}$ と $\boldsymbol{x}^{T}\boldsymbol{A}\boldsymbol{x}$ の2つの形式が頻出する。

- 前者は**線形結合**（もしくは一次結合）、後者は**二次形式**と呼ぶ。

これらはスカラーにおける一次式（$ax+b$）や二次式（$ax^2+bx+c$）をベクトルに拡張した形と理解できる。

線形結合の計算例：
$$
\begin{aligned}
\boldsymbol{b}&=\begin{bmatrix}
1 \\
2\end{bmatrix},\
\boldsymbol{x}=\begin{bmatrix}
x_{1} \\
x_{2}
\end{bmatrix}\\
\boldsymbol{b}^{T}\boldsymbol{x}&=\begin{bmatrix}
1 & 2
\end{bmatrix}\begin{bmatrix}
x_{1} \\
x_{2}\end{bmatrix}=x_{1}+2x_{2}\end{aligned}
$$

$\boldsymbol{x}$ の各要素 $x_{1}, x_{2}$ に関して一次式となっていることがわかる。

二次形式の計算例：

$$
\begin{aligned}
\boldsymbol{A}&=
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix},\  
\boldsymbol{x}=
\begin{bmatrix}
x_{1} \\
x_{2}
\end{bmatrix}\\
\boldsymbol{x}^{T}\boldsymbol{A}\boldsymbol{x}&=
\begin{bmatrix}
x_{1} & x_{2}
\end{bmatrix}
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
\begin{bmatrix}
x_{1} \\
x_{2}
\end{bmatrix}\\
&=x^{2}_{1}+5x_{1}x_{2}+4x_{2}^{2}
\end{aligned}
$$

各要素の二次式となっている。したがって、任意の二次関数は、$c$ をスカラーの定数項とすると以下のように表現できる。
$$
\boldsymbol{x}^{T}\boldsymbol{A}\boldsymbol{x} + \boldsymbol{b}^{T}\boldsymbol{x} + c
$$


### ベクトルによる微分と勾配

多変数の関数をベクトルで微分することを考える。
関数の各成分をベクトルの各成分で偏微分し、それらを並べてベクトル（または行列）にしたものを**勾配**（Gradient）と呼ぶ。

例えば、線形結合 $\boldsymbol{b}^{T}\boldsymbol{x}$ をベクトル $\boldsymbol{x}$ で微分することを、以下のように表す。

$$
\dfrac {\partial }{\partial \boldsymbol{x}}( \boldsymbol{b}^{T}\boldsymbol{x})
$$

具体的に $\boldsymbol{b} = [3, 4]^T$ の場合：

$$
\begin{aligned}
\dfrac {\partial }{\partial \boldsymbol{x}}( \boldsymbol{b}^{T}\boldsymbol{x}) &=\dfrac {\partial }{\partial \boldsymbol{x}}( 3x_{1}+4x_{2}) \\
&=\begin{bmatrix}
\dfrac {\partial }{\partial x_{1}} ( 3x_{1}+4x_{2})  \\
\dfrac {\partial }{\partial x_{2}} ( 3x_{1}+4x_{2})
\end{bmatrix}
=\begin{bmatrix} 3 \\ 4 \end{bmatrix} = \boldsymbol{b}
\end{aligned}
$$

このように、スカラーの微分と同様に $\boldsymbol{x}$ の係数が結果として得られる（微分の定義により結果が行ベクトルになるか列ベクトルになるかの慣習「レイアウト」の違いがあるが、機械学習では勾配を元のベクトルと同じ形状で扱うことが多い）。

重要な微分の公式：
$$
\begin{aligned}
( 1) &\ \dfrac {\partial}{\partial \boldsymbol{x}}( \boldsymbol{c} ) = \boldsymbol{0}\\
( 2) &\ \dfrac {\partial }{\partial \boldsymbol{x}}( \boldsymbol{b}^{T}\boldsymbol{x}) = \boldsymbol{b}\\
( 3) &\ \dfrac {\partial }{\partial \boldsymbol{x}}( \boldsymbol{x}^{T}\boldsymbol{A}\boldsymbol{x}) =( \boldsymbol{A}+\boldsymbol{A}^{T}) \boldsymbol{x}\end{aligned}
$$

※ $\boldsymbol{A}$ が対称行列（$\boldsymbol{A} = \boldsymbol{A}^T$）であれば、$(3)$ は $2\boldsymbol{A}\boldsymbol{x}$ となり、スカラーの $ax^2$ の微分 $2ax$ との類似性がより明確になる。

複雑な微分に関しては、[MatrixCookBook](http://class.west.sd.keio.ac.jp/dataai/text/matrixcookbook.pdf) などのリファレンスを参照するとよい。


## ラグランジュの未定乗数法
制約条件 $g(\boldsymbol{x})=0$ のもとで関数 $f(\boldsymbol{x})$ を最大化・最小化したい等式制約付き最適化問題において、ラグランジュ関数を以下のように定義する。

$$L(\boldsymbol{x},\lambda)=f(\boldsymbol{x})-\lambda g(\boldsymbol{x})$$

このとき、極値を与える候補は、以下の条件を満たす点 $\boldsymbol{x}$ である。
$$\dfrac{\partial L}{\partial x_1}=\dfrac{\partial L}{\partial x_2}=\cdots=\dfrac{\partial L}{\partial x_n}=\dfrac{\partial L}{\partial\lambda}=0$$

この方法により変数が1つ増えるが、制約式も方程式に組み込まれるため、連立方程式として解くことが可能になる。
機械学習では、サポートベクターマシン（SVM）の最適化や、情報理論におけるエントロピー最大化などで利用される。

**例題**：$x^2+y^2=1$ のもとで $f(x,y)=2x+3y$ を最大化せよ。

**解**：
$$L(x,y,\lambda)=2x+3y-\lambda(x^2+y^2-1)$$
1. $\tfrac{\partial L}{\partial x}=2-2x\lambda=0 \Rightarrow x = 1/\lambda$
2. $\tfrac{\partial L}{\partial y}=3-2y\lambda=0 \Rightarrow y = 3/(2\lambda)$
3. $\tfrac{\partial L}{\partial \lambda}=-(x^2+y^2-1)=0$

3に1, 2を代入すると $(1/\lambda)^2 + (3/2\lambda)^2 = 1 \Rightarrow 13/(4\lambda^2) = 1 \Rightarrow \lambda = \pm\sqrt{13}/2$
これより $(x,y)=\left( \pm\dfrac{2}{\sqrt{13}}, \pm\dfrac{3}{\sqrt{13}} \right)$ (複号同順) が候補となり、正の方が最大値 $\sqrt{13}$ を与える。


## ヤコビ行列（ヤコビアン）

多入力多出力関数の微分であるヤコビ行列は、ニューラルネットワークの誤差逆伝播法（Backpropagation）を理解する上で重要である。

深層学習の文脈（特に誤差逆伝播法）では、各層の出力に対する入力や重みの勾配を計算する。ベクトル値関数 $\boldsymbol{y} = \boldsymbol{Ax}$ の入力 $\boldsymbol{x}$ に対する勾配は $\boldsymbol{A}^T$ となり（分母レイアウトの場合）、これがヤコビ行列の転置に相当するため、計算過程で転置行列が頻繁に現れる。

### 状況設定

$\boldsymbol{x} = (x_1, x_2, \cdots , x_n)^T$ を入力として
$\boldsymbol{y} = (y_1, y_2, \cdots , y_m)^T$ が出力される $m$ 次元ベクトル値関数を考える。
各出力成分 $y_i$ が各入力成分 $x_j$ で偏微分可能であれば、それらを要素とする行列を構成できる。

### 定義

#### ヤコビ行列 (Jacobian Matrix)

各要素を $J_{ij} = \frac{\partial y_i}{\partial x_j}$ とする $m\times n$ 行列 $\mathbf{J}$ をヤコビ行列という。

例：$m=2, n=2$ の場合
$$
J = \begin{pmatrix}\frac{\partial y_1}{\partial x_1} & \frac{\partial y_1}{\partial x_2} \\
\frac{\partial y_2}{\partial x_1} & \frac{\partial y_2}{\partial x_2}
\end{pmatrix}
$$

#### ヤコビ行列式 (Jacobian Determinant)

正方行列（$m=n$）であるヤコビ行列の行列式 $|J|$ をヤコビ行列式と呼ぶ。
- 一般に「ヤコビアン」という用語は、ヤコビ行列を指すこともあれば、その行列式を指すこともあるが、多重積分における変数変換の文脈では行列式を指すことが一般的である。

ヤコビ行列式は変換による空間の「拡大率」を表し、重積分における変数変換式などで用いられる。
$$
\iint_D f(x, y) dxdy
= \iint_D f(\varphi(u, v)) \left| \frac{\partial (x, y)}{\partial (u, v)} \right| dudv
$$

### ヤコビ行列の意味

- **1変数の場合**：微分係数は接線の傾きであり、関数の1次近似（線形近似）を表す。
  $y(x) \simeq y(x_0) + y'(x_0)(x-x_0)$

- **多変数の場合**：ヤコビ行列がこの「傾き」に相当し、ベクトル値関数の1次近似を与える。
  $$\boldsymbol{y}(\boldsymbol{x}) \simeq \boldsymbol{y}(\boldsymbol{x_0}) + \mathbf{J}(\boldsymbol{x_0})(\boldsymbol{x}-\boldsymbol{x_0})$$

## 確率

機械学習のモデル学習やパラメータ推定の根底には確率論がある。

$x$ を**確率変数** (stochastic variable / random variable) としたとき、確率 $p(x)$ について以下の関係が成り立つ。

$$
\begin{align}
\sum_{x } p(x) &= 1 \quad (または \int p(x)dx = 1) \\
p(x) & \geq  0
\end{align}
$$

**同時確率** (joint probability)：2つの事象が同時に起きる確率 $p(x, y)$。

**周辺化** (marginalization)：同時確率から一方の変数の和（または積分）をとって消去する操作。注目した確率変数の確率が得られる。
$$
p(x) = \sum_y p(x, y)
$$

**条件付き確率** (conditional probability)：事象 $x$ が起きたという条件下での事象 $y$ の確率 $p(y|x)$。
$$
p(y|x) = \frac{p(x, y)}{p(x)}
$$

これらより得られる以下の関係式を**ベイズの定理**と呼ぶ。
$$
p(x|y) = \frac{p(y | x)p(x)}{p(y)}
$$
- $p(x)$ を**事前確率**（データを得る前の知識）、$p(x|y)$ を**事後確率**（データ $y$ を得た後の知識）と呼ぶ。
- $p(y|x)$ は**尤度**（パラメータ $x$ のもとでデータ $y$ が得られる尤もらしさ）である。

**例：火事と煙**
- 火事が起こる確率 $P(Fire)=0.01$、煙が上がっている確率 $P(Smoke)=0.1$ とする。
- 火事の際に煙が上がる条件付き確率（尤度）を $P(Smoke|Fire)=0.9$ とする。
- 「煙が見えたときに実際に火事である確率（事後確率）」は、ベイズの定理より：
  $P(Fire|Smoke)=\frac{0.01 \times 0.9}{0.1} = 0.09 \quad (9\%)$

直感的には煙＝火事と思いがちだが、火事自体の頻度が低ければ、事後確率はそれほど高くならないことが理論的に示される。初期のスパム判定フィルタも、このベイズ理論を応用した「ナイーブベイズ分類器」が主流であった。

## 尤度と最尤推定

パラメータ $\theta$ を持つ確率モデル $p(x; \theta)$ において、実際に観測されたデータ $X$ がそのモデルから生成されたと仮定したときの確率（または密度）の値を **尤度** (Likelihood) と呼ぶ。

**最尤推定** (Maximum Likelihood Estimation, MLE) は、観測されたデータ $X$ を最も生成しそうな（尤度を最大にする）パラメータ $\theta$ を求める手法である。

観測データ $X = \{x^{(1)}, \ldots, x^{(N)}\}$ が独立同一分布（i.i.d.）に従う場合、全体の尤度は個別の尤度の積となる。
$$
L(\theta) = \prod_{i=1}^N p(x^{(i)}; \theta)
$$

積の形式は計算が煩雑になり、値が非常に小さくなる（アンダーフロー）ため、通常は対数をとった **対数尤度** を用いる。対数関数は単調増加であるため、尤度の最大化と対数尤度の最大化は等価である。
$$
\log L(\theta) = \sum_{i=1}^N \log p(x^{(i)}; \theta)
$$

**例：コイン投げ**
- 表が出る確率 $\theta$ を推定したい。10回投げて表(1)が5回、裏(0)が5回出たとする。
- $L(\theta) = \theta^5 (1-\theta)^5$
- 対数尤度を $\theta$ で微分して 0 とおくと、$\theta = 0.5$ が得られる。観測データにおいて各事象が同数出現する場合、表裏の確率が等しいと推定するのが最も「尤もらしい」ということになる。

最小二乗法による回帰分析も、ノイズが正規分布に従うと仮定した場合の最尤推定と数学的に等価であることが知られている。

### 事後確率最大化推定 (MAP推定)

最尤推定は観測データのみに依存するが、データ数が少ない場合に極端な推定結果（過学習）になりやすい欠点がある。例えば、コインを2回投げてたまたま2回とも表だった場合、最尤推定では「表が出る確率は100%」と結論づけてしまう。

これに対し、パラメータ $\theta$ 自体も何らかの確率分布（事前分布）に従うと考え、事前知識を考慮してパラメータを推定するのが **事後確率最大化 (Maximum A Posteriori, MAP) 推定** である。

ベイズの定理より事後確率は以下に比例する。
$$p(\theta|X) \propto p(X|\theta) p(\theta)$$

MAP推定はこの事後確率を最大化する $\theta$ を求める。これは、最尤推定の目的関数に対数事前分布 $\log p(\theta)$ というペナルティ項を加えたものと解釈できる。機械学習における**正則化**（L2正則化など）は、パラメータが特定の値（例えば0付近）に集まりやすいという事前分布を仮定したMAP推定に対応している。

ここで、具体的な例を使いながら理解を深める。

藤井聡太 氏（現・八冠）はデビュー以来29連勝という驚異的な記録を達成した。これを単純な確率（最尤推定）で考えると、勝率 $\theta$ が $1$ であるときが最も尤もらしいことになるが、現実的ではない。

データの数が少ない（例えば4戦4勝）とき、たまたまである可能性も高く、だからといって勝率1とは言いにくい。ここでMAP推定を用いる。

事前分布として、勝率の議論でよく使われる **ベータ分布** を用いる。ベータ分布 $\text{Beta}(\theta|\beta, \beta)$ を仮定し、$n$ 回中 $v$ 回勝利したときのMAP推定値は以下のようになる。
$$\theta_{MAP}=\frac{v+\beta-1}{n+2(\beta-1)}$$

例えば、$\beta=2$ と設定（ラプラススムージングに近い考え方）すると、29連勝した際の推定勝率は約 $0.968$ となり、100%とは言わずとも極めて高い実力が示唆される一方、偶然の要素も考慮した控えめな推定となる。ハイパーパラメータ $\beta$ を大きくするほど、事前知識（「勝率は0.5付近のはずだ」という信念）の重みが強まり、観測データによる偏りが抑制される。

### ベイズ推定
MAP推定は事後確率分布の「点（モード）」のみを見るが、**ベイズ推定**は事後分布そのものを利用し、パラメータの不確実性を考慮する。データ数が増えるほど分布は鋭くなり、推定の「確信度」が高まっていく。連勝を重ねるほど、$\theta$ の確率分布が $1$ の方向に移動しつつ、幅が狭まっていく様子を可視化することで、実力に対する確信が深まるプロセスを理解できる。

### 統計量：分散と標準偏差

データのばらつきを評価する指標として**分散** $\sigma^2$ がある。

**標本分散**：観測データそのものの平均からのばらつき。
$$\sigma ^{2}=\dfrac {1}{N}\sum ^{N}_{n=1}\left( x_{n}-\overline {x}\right) ^{2}$$

**不偏分散**：標本から母集団の分散を推定する際に用いられ、$N-1$ で割ることでバイアスを補正する。
$$\hat{\sigma} ^{2}=\dfrac {1}{N-1}\sum ^{N}_{n=1}\left( x_{n}-\overline {x}\right) ^{2}$$

データ解析において、手元のデータが「母集団そのもの」なのか、より大きな集団から抽出された「標本」なのかを意識することが重要である。サンプル数 $N$ が十分に大きければ両者の差は無視できるが、小規模データでは不偏分散を用いるのが標準的である。

### 正規分布

**正規分布**（ガウス分布）は自然界や統計データで最も頻繁に現れる分布である。

- 中心極限定理により、独立な多数の因子の和として現れる現象は正規分布に近づく。
- 平均 $\mu$、標準偏差 $\sigma$ としたとき、$\mu \pm 2\sigma$ の範囲に約 95.4%、$\mu \pm 3\sigma$ の範囲に約 99.7% のデータが含まれる。

必ずしもデータが正規分布に従うとは限らないが、多くの機械学習アルゴリズム（線形回帰、LDAなど）は、データの背後に正規分布やその誤差を仮定している。分析の際は、ヒストグラムなどで分布を可視化し、仮定が妥当か確認する習慣をつけるべきである。

### 標準偏差を利用したスケーリング

機械学習の前処理として、各特徴量のスケール（単位や桁数）を揃える「スケーリング」は非常に重要である。例えば、身長(m)と年収(円)をそのまま距離計算に用いると、値の大きい年収のみが結果に支配的な影響を与えてしまう。

- **Min-Maxスケーリング**（正規化）：データを [0, 1] の範囲に収める。
  $$\widetilde{x} = \dfrac{x - x_{\min}}{x_{\max} - x_{\min}}$$
- **標準化** (Standardization)：平均 0、標準偏差 1 に変換する。外れ値の影響を比較的受けにくい。
  $$\widetilde{x}  = \dfrac{x - \bar{x}}{\sigma}$$


## データの分割と汎化性能

機械学習の究極の目的は、学習に使ったデータ（訓練データ）を完璧に記憶することではなく、未知のデータに対して正しく予測できる能力、すなわち**汎化性能**を獲得することである。

### 訓練・検証・テストデータの分割

モデルの性能を客観的に評価するためには、データを以下の3つに分割して管理することが現代の機械学習における標準的なプラクティスである。

1.  **訓練データ (Training Data)**
    - モデルの内部パラメータ（重みやバイアス）を学習させるために使用するデータ。
2.  **検証データ (Validation Data)**
    - 学習の過程で、モデルの汎化性能をチェックするために使用するデータ。
    - ハイパーパラメータ（学習率、層の数、正則化の強さなど）の調整や、どのモデルを採用するかの決定（モデル選択）に用いられる。
3.  **テストデータ (Test Data)**
    - 全ての学習およびチューニングが完了した後に、最終的な性能を測定するために一度だけ使用するデータ。
    - 開発の過程でこのデータを見てしまうと、「テストデータに特化した調整」を行ってしまう可能性があり（データリーク）、正当な評価ができなくなる。

#### よくある誤り：検証データとテストデータの混同
「テストデータで精度が上がらなくなったので学習率を変えた」といった操作を繰り返すと、それは実質的にテストデータを使って学習していることになり、真の未知データに対する性能を見誤ることになる。この「検証」と「テスト」の厳密な分離が、信頼性の高いシステム構築の第一歩である。

### 汎化性能に関する現代的な視点

#### バイアス・バリアンス・トレードオフ（古典的視点）
モデルが単純すぎるとデータの特徴を捉えきれず（高バイアス・未学習）、複雑すぎるとノイズまで学習してしまう（高バリアンス・過学習）というトレードオフの関係がある。このため、適切な複雑さのモデルを選択することが重要とされてきた。

#### 二重降下現象 (Double Descent) と過剰パラメータ化（現代的視点）
近年のディープラーニングの研究では、古典的な理論では説明できない現象が観察されている。モデルのパラメータ数を増やしていくと、一度過学習によりテスト誤差が増大するが、さらにパラメータ数を増やして「モデルが訓練データを完全に再現（補間）できる」領域を超えると、再びテスト誤差が減少し始めるという現象であり、**二重降下現象**と呼ばれる。

現在のトレンドは、巨大なモデル（過剰パラメータ化されたモデル）を使いつつ、適切な正則化や最適化アルゴリズムを用いることで、訓練誤差をゼロにしながらも高い汎化性能を実現する方向にある。これは、巨大なモデルほど「滑らかな」関数を学習しやすいといった特性に起因すると考えられており、LLM（大規模言語モデル）の成功を支える理論的背景の一つとなっている。
